```python
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
whales_navs_csv = pd.read_csv(
    Path('esg_082122.csv'),
    index_col="Datetime",
    parse_dates=True,
    infer_datetime_format=True
)
whales_navs_csv.head()
```

```python
df=whales_navs_csv['Close']
```

```python
daily_returns = df.pct_change().dropna()
daily_returns.head()
```

```python
daily_returns.plot.line(subplots=True, figsize=(15,9), title="Daily Returns")
```

```python
cumulative_returns = (1+ daily_returns).cumprod()
cumulative_returns.head()
```

```python
cumulative_returns.plot(figsize=(20,9), title="Cumulative Returns")
```
